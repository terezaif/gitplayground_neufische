```python
import pandas as pd

url="https://raw.githubusercontent.com/oem/Hamburg.jl/master/src/covid-19/agegroups.csv"
age_groups=pd.read_csv(url)
```

```python
#adding a date column
age_groups["recorded_date"]= pd.to_datetime(age_groups.recordedat) 
age_groups.info()
```

```python
import seaborn as sns
sns.scatterplot(data=age_groups, x="recorded_date", y="male", style="age")
```

```python
age_groups.drop("recordedat",axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,14))
sns.lineplot(data=age_groups, x= "recorded_date", y="male", style="age", color="blue")
plot2=sns.lineplot(data=age_groups, x= "recorded_date", y="female", style="age", color = "red")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel("cases")
locs, labels = plt.xticks()
plt.setp(labels, rotation=45);
```
